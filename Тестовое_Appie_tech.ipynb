Создадим тестовые данные

```python
import numpy as np
import pandas as pd
import random
import datetime
#фиксация_времени
np.random.seed(42)
random.seed(42)


#функция генерации дат
def generate_date():
  year = random.randint(2022, 2023)
  month = random.randint(1, 12)
  day = random.randint(1, 28)
  return datetime.date(year, month, day).strftime('%d.%m.%Y')


n=100
costs = {'adCost':np.random.exponential(100,size = n)+1,
         'utm_term': random.choices(['a','b','c','d','e','f'], k=n),
         'utm_campaign':random.choices(['A', 'B', 'C'],k=n),
        'utm_medium':random.choices(['a','b','c'], k=n),
        'utm_source':random.choices(['google', 'yandex', 'facebook'],k=n),
        'date':[generate_date()for _ in range(n)]
                                       }
df1 = pd.DataFrame(costs)
#конвертация даты в дату
df1['date'] = pd.to_datetime(df1['date']).dt.date
```

```python
#функция гереации id из символов
import numpy as np
import random
import pandas as pd
def generate_user_id():
  charrete = 'abcdefj123456789'
  return ''.join(random.choice(charrete) for _ in range (15))

n=100
sessions = {'datetime':np.random.exponential(100,size = n)+1,
         'ClientID': [generate_user_id() for _ in range(n//2)]*2,
         'sessionID': np.random.randint(1,11,size = n)

                                       }
df3 = pd.DataFrame(sessions)
df3 = df3.join(df2)
```
